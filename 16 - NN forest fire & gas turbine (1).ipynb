# Neural Network(Forest Fire)

```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
forest =pd.read_csv("C:/Users/Hp/Downloads/forestfires.csv")
```

```python
forest.head()
```

```python
forest.shape
```

```python
forest.info()
```

```python
forest.loc[forest.size_category=="large","size_category"] = 1
forest.loc[forest.size_category=="small","size_category"] = 0
```

```python
forest.size_category.value_counts()
```

```python
forest['month']=forest['month'].astype('category')
forest['day']=forest['day'].astype('category')
forest['size_category']=forest['size_category'].astype('int')
```

```python
forest.dtypes
```

```python
from sklearn import preprocessing                      
label_encoder = preprocessing.LabelEncoder()
```

```python
forest['month']=label_encoder.fit_transform(forest['month'])
forest['day']=label_encoder.fit_transform(forest['day'])
```

```python
forest
```

```python
# Splitting the data into x and y as input and output
x = forest.iloc[:,0:30]
y = forest.iloc[:,30]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Splitting the data into training and test dataset

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=0)
```

```python
x_train, x_test, y_train, y_test 
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
```

```python
# Now apply the transformations to the data:
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(30,30))
```

```python
mlp.fit(x_train,y_train)
```

```python
prediction_train=mlp.predict(x_train)
prediction_test = mlp.predict(x_test)
```

```python
prediction_test
```

```python
prediction_train
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,prediction_test))
np.mean(y_test==prediction_test)
np.mean(y_train==prediction_train)
```

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
```

```python
# create model
model = Sequential()
model.add(Dense(12, input_dim=30, activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
# evaluate the model
scores = model.evaluate(x,y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

# Neural Networking (Gas Turbine)

```python
#Import Libraries
import pandas as pd 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.distplot(a=gas.TEY)
```

```python
X=gas.drop('TEY',axis=1)
X
```

```python
Y=gas['TEY']
Y
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(X)
x_scaled
```

```python
scaled_gasturbines_data=pd.DataFrame(x_scaled,columns=X.columns)
scaled_gasturbines_data
```

```python
X_transformed=scaled_gasturbines_data
X_transformed
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_transformed,Y,test_size=0.20,random_state=123)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
model=Sequential()
model.add(Dense(units=10,input_dim=10,activation ='relu',kernel_initializer='normal'))
model.add(Dense(units=6,activation='tanh',kernel_initializer='normal'))
model.add(Dense(units=1,activation='relu',kernel_initializer='normal'))
```

```python
model.compile(optimizer='adam',loss='mean_absolute_error',metrics=['mse'])
```

```python
model.fit(X_train,Y_train, epochs=100, batch_size=20)
```

```python
scores = model.evaluate(X_test,Y_test)
print((model.metrics_names[1]))
```
